#DATA VISUALISATION

*Data Visualization is the process of presenting data in the form of graphs or charts. It helps to understand large and complex amounts of data very easily.

*It allows the decision-makers to make decisions very efficiently and also allows them in identifying new trends and patterns very easily.

*It is also used in high-level data analysis for Machine Learning and Exploratory Data Analysis (EDA).  Data visualization can be done with various tools like Tableau, Power
BI,Python.

#MATPLOTLIB

*Matplotlib is a low-level library of Python which is used for data visualization.

*It is easy to use and emulates MATLAB like graphs and visualization.

*This library is built on the top of NumPy arrays and consist of several plots like line chart, bar chart, histogram, etc.

*It provides a lot of flexibility but at the cost of writing more code.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=[1,2,3]
y=[4,5,6]
plt.plot(x,y) #(1,4)(2,5)(3,6)
plt.grid()
plt.show()
```

```python
#Univariate - Numerical
```

```python
data = {
    "Salary" : [25000, 30000, 37000, 28000, 39000, 48000, 55000, 52000, 35000, 26000, 27000, 31000, 33000, 29000, 40000, 36000, 34000, 45000, 47000, 50000]
}
df=pd.DataFrame(data)
print(df.head())

print(df.shape)
```

```python
#Line plot:

plt.plot(df["salary"], color="red", marker="*", linestyle=":", linewidth=2)
plt.grid()
plt.show()
```

```python

#Histogram  - frequency of the data distribution

plt.hist(df["Salary"],bins=5)
plt.show()
```

```python
#Boxplot:

plt.boxplot(df["Salary"])
plt.show()
```

```python
df.loc[20] = [0]

plt.boxplot(df["Salary"])
plt.show()
```

```python
#Univarite : categorical
```

```python
import pandas as pd

df = pd.DataFrame({"salary": [25000, 30000, 35000, 40000, 45000, 46000, 47000, 48000, 49000, 50000]})
df["dept"] = ['HR', 'IT', 'Finance', 'HR', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR']
print(df.head())
```

```python
# Pie chart:

count = df["dept"].value_counts()    #counts the unique value indept
count
```

```python
plt.pie(count, labels = count.index, autopct= "%1.2f", explode = [0, 0.1, 0.2])
plt.axis("equal")    #makes as a complete circle
plt.show()

#labels - names of the each pie
#autopct - to show the values of each pie  (1.0f- just the number)
#explode - spaces btw each pie
```

```python
#countplot (barchar)

plt.bar(count.index,count,color=["green","black","red"])
plt.show()
```

```python
df.head()
```

```python
df["Age"] = [22, 25, 29, 24, 30, 35, 40, 36, 28, 22]
df.head()
```

```python
#scatter plot

plt.scatter(df["Age"], df["salary"], color = "orange")
plt.show()
```

```python
#Line plot

plt.plot(df["Age"], df["salary"], color = "orange")
plt.show()
```

```python
#Bar chart

plt.bar(df["Age"], df["salary"], color = "green")
plt.show()
```

```python
#Bivariate : NUmerical - categorical
```

```python
hr_sal = df[df["dept"] == "HR"]["salary"]
hr_sal
```

```python
it_sal = df[df["dept"] == "IT"]["salary"]
it_sal
```

```python
finance_sal = df[df["dept"] == "Finance"]["salary"]
finance_sal
```

```python
#boxplot

plt.boxplot([hr_sal, it_sal, finance_sal])
plt.xticks([1, 2, 3], ['HR', 'IT', 'Finance'])
plt.show()
```

```python
#pie chart

salary_by_dept= df.groupby("dept")["salary"].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept, labels = salary_by_dept.index, autopct= "%1.2f")
plt.axis("equal")
plt.show()

#shadow - to get lil 3d effect
```

```python
#barplot

hr_mean=sum(hr_sal)/len(hr_sal)
it_mean=sum(it_sal)/len(it_sal)
finance_mean=sum(finance_sal)/len(finance_sal)
print(hr_mean, it_mean, finance_mean)
```

```python
plt.bar(["HR", "IT", "Finance"], [hr_mean, it_mean, finance_mean])
plt.show()
```

```python
df["experience"]=[1, 2, 4, 1.5, 5, 6, 3, 7, 2.5, 4.5]
df.head()
```

```python
#bubble plot

plt.scatter(df["Age"], df["salary"], s=df["experience"]*100)
plt.show()
```

```python
# 2 numerical and 1 categorical column:
```

```python
plt.scatter(df["Age"], df["salary"], c = df["dept"].map({"HR": "yellow", "IT":"blue", "Finance": "orange"}))
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary vs Dept")
plt.legend()
plt.show()
```

```python
color = {"HR": "yellow", "IT":"blue", "Finance": "orange"}

for dept, color in color.items():
    df_dept = df[df["dept"] == dept]
    plt.scatter(df_dept["Age"], df_dept["salary"], c = color, label = dept)

plt.legend()
```
