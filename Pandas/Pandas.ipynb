Data structures are collection of data types that provide the best way of organizing items(values) in terms of memory usage.

Primary Data structures

*Series: A one-dimensional labeled homogeneous array(of the same datatype).size immutable.

*Dataframe: A two dimensional labeled,size-mutable tabular structure with potentially heterogeneously typed columns.

```python
fruit_protein = {
    "Avocado": 2.0,       # grams of protein
    "Guava": 2.6,
    "Blackberries": 2.0,
    "Oranges": 0.9,
    "Banana": 1.1,
    "Apples": 0.3,
    "Kiwi": 1.1,
    "Pomegranate": 1.7,
    "Mango": 0.8,
    "Cherries": 1.0
}
```

```python
s2 = pd.Series(fruit_protein, name = "Protein")
s2
```

```python
# Conditional Selection:

s2[s2>1]   #to get in boolean values use s2>1
```

```python
# Logical Operators: and, or, not

s2[(s2>0.5) | (s2 <= 2)]
```

```python
# not operation:

s2[~(s2>1)]
```

#DATA CLEANING


```python


df.isnull().sum()
```

```python
df.dropna(how = "any") # any row that had any null value
```

```python
df.dropna(how = "all") #if all the values in any row are null then we drop that row
```

```python
df["Name"].replace("charlie","Rose")
```

```python
df["Name"]=df["Name"] = df["Name"].replace("charlie","Rose")   #changes in the original dataset as well
df
```

#DUPLICATE VALUES

```python
#Duplicates

df_dup=df[df.duplicated()]  #can add keep="first" - shows the last duplicated value
df_dup
```

```python
#to delete the duplicated value

df_dup=df[df.duplicated(keep="last")]
df_dup
```

```python
#Invalid values
#Lambda ->python

df["Promoted Salary"]=df["Promoted Salary"].apply(lambda x:x/10 if x>650000 else x)
df
```

```python
#name="alice_fernandes" #to change the name which has first and last name
#df[["first_name","last_name"]] = df["name"].str.split("_")
```

```python
#apply and lambda

def multiplying_age(x):
   return x*2
df["Age"] = df["Age"].apply(multiplying_age)
df
```

```python
df["Age"] = df["Age"].apply(lambda x:x/2)

df    #back to the original values
```

```python
# Joins and Merges

department_info = {
    "Dept": ["HR", "IT", "Finance"],
    "Location": ["New York", "San Francisco", "Chicago"],
    "Manager": ["Laura", "Steve", "Nina"]
}

df2 = pd.DataFrame(department_info)
df2
```

```python
# Concat

pd.concat([df, df2])   #adds 2 dataframes
```

```python
pd.concat([df, df2], axis = 1)
```

```python
pd.merge(df, df2, on = "Dept")   #dept appears only once
```
