```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    0: [1.0, 2.0, np.nan, np.nan, np.nan],
    1: [np.nan, 3.0, 4.0, 8.0, np.nan],
    2: [2.0, 5.0, 6.0, np.nan, np.nan],
    3: [np.nan, np.nan, np.nan, np.nan, np.nan]
})
print(df)
```

```python
df_fill=df.ffill(axis=1)   #Forward fill across columns means fill NaN horizontally (left to right)
print(df_fill)
```

```python
df.dropna(axis=1, how='all', inplace=True)
print(df)
```

```python
df.dropna(axis=0,how='all',inplace=True)
print(df)
```

```python
print(df[df.count(axis=1)>=2])
```
